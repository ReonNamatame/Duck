```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_acce = pd.read_csv('accelerometer.csv', header=None)
df_acce.columns = ['x_axis', 'y_axis', 'z_axis']

df_acce.head()
```

```python
df_dist = pd.read_csv('distance.csv', header=None)
df_dist.columns = ['distance']

df_dist.head()
```

```python
plt.figure(figsize=(30, 15))
plt.plot(range(0, 20), df_acce['x_axis'], color='green', label='x axis')
plt.plot(range(0, 20), df_acce['y_axis'], color='red', label='y axis')
plt.plot(range(0, 20), df_acce['z_axis'], color='yellow', label='z axis')
plt.legend(loc="upper right")
plt.xlabel("time (second)")
plt.ylabel("Acceleration")
plt.show()
```

```python
plt.figure(figsize=(30, 15))
plt.plot(range(0, 20), df_dist['distance'], color='green')
plt.xlabel("time (second)")
plt.ylabel("Distance (cm)")
plt.show()
```
